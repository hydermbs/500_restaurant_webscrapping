```python
web_data = pd.read_csv('Casual Dine In Restaurant.csv')
web_data
```

```python
#function to iterate the restaurant, whether their website is functional or not.
from datetime import datetime
import hashlib


def get_content_hash(url):
    try:
        response = requests.get(url)
        content = response.content

        # Calculate the MD5 hash of the content
        content_hash = hashlib.md5(content).hexdigest()
        return content_hash
    except requests.exceptions.RequestException as e:
        print(f"Error for {url}: {e}")
        return None

```

```python
website_status=[]
for website in web_data['Website']:
    print(f"\nChecking {website}:")
    
    # Check Content Hash
    content_hash = get_content_hash(website)
    if content_hash:
        print(f"Content MD5 Hash: {content_hash}")
        website_status.append('True')
    else:
        website_status.append('False')
    
```

```python
#update website status list in Data frame
web_data['website_status']=website_status
web_data
```

```python
#export list of customer without website
mid_pot_customer = web_data[web_data['website_status']=='False']
mid_pot_customer.to_csv('Customer_wo_Website.csv')
```

```python
#Create new data set to identify whether they online ordering or not.
online_web = web_data[web_data['website_status']=="True"]
```

```python
def identify_keywords_presence(urls, keywords):
    result_list = []

    for url in urls:
        try:
            response = requests.get(url)
            soup = BeautifulSoup(response.text, 'html.parser')

            # Check if the page contains any of the specified keywords in the navigation bar
            navigation_bar = soup.find('ul')  

            if navigation_bar:
                list_items = navigation_bar.find_all('li')
                if list_items is not None and any(keyword.lower() in item.text.lower() for keyword in keywords for item in list_items):
                    result_list.append(True)
                else:
                    result_list.append(False)
            else:
                result_list.append(False)

        except requests.exceptions.RequestException as e:
            print(f"Error for {url}: {e}")
            result_list.append(False)

    return result_list
```

```python
websites = online_web['Website']
words = ['order','online','shop']
results = identify_keywords_presence(websites,words)
```
